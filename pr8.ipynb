Импортируем библиотеки.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import KFold
```

Загружаем датасет.

```python
df = pd.read_csv("vgsales4.csv")
df
```

Преобразовываем категариалбные значения в бинарные.

```python
OE = OrdinalEncoder()
df[['Name', 'Platform', 'Year', 'Publisher']] = OE.fit_transform(df[['Name', 'Platform', 'Year', 'Publisher']])
OHE = OneHotEncoder()
df['Genre'] = OHE.fit_transform(df[['Genre']]).toarray()
df
```

Мультиклассовая классификация.

К-блочная кросс-валидация - это процедура, которая используется для оценки производительности моделей машинного обучения на новых данных. Она является популярным методом для сравнения и выбора моделей в задачах прогнозирования, так как он прост в понимании и реализации.

Основные параметры, используемые в к-блочной кросс-валидации, включают:

n_splits=5: Количество блоков или фолдов, на которые разбиваются данные. В каждом фолде модель обучается на остальных фолдах и оценивается на текущем фолде. Это позволяет использовать каждый фолд один раз в качестве тестового набора данных.

shuffle=False: Определяет, следует ли перемешивать данные перед разбиением на блоки. Если shuffle=True, то данные случайным образом перемешиваются перед разбиением, что помогает уменьшить зависимость результатов от исходного порядка данных.

random_state=None: Управляет случайностью разбиения на блоки при shuffle=True. Задание фиксированного значения random_state обеспечивает повторяемость разбиений на блоки при повторном запуске кода с тем же значением random_state.

```python
Y = df["Genre"]
X = df.drop(columns = "Genre")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

kf.split(df) выполняет разбиение DataFrame df на обучающие и тестовые наборы данных с использованием к-блочной кросс-валидации.

```python
kf.split(df)
```

Импорт библиотек и создание моделей для каждой библиотеки.

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

Выполняется кросс-валидация с использованием разбиения.

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]

    gnb.fit(X_train_kfold,Y_train_kfold)
    predict_gnb = gnb.predict(X_test_kfold)

    knc.fit(X_train_kfold,Y_train_kfold)
    predict_knc = knc.predict(X_test_kfold)

    dtr = dtr.fit(X_train_kfold,Y_train_kfold)
    predict_dtr = dtr.predict(X_test_kfold)
```

Вывод результатов.

```python
print("Набор предсказаний Наивного байесовского классификатора: ",predict_gnb)
print("Набор предсказаний К-ближайших соседей: ",predict_knc)
print("Набор предсказаний регрессии дерева решений: ",predict_dtr)
```

Регрессия – группа Моделей, используемых для прогнозирования непрерывных значений, таких как цены на недвижимость с учетом их характеристик (размер, цена и т.д.).
Далее делаем тоже самое только с числовой переменной и другими библиотеками.

```python
Y = df["Global_Sales"]
X = df.drop(columns = "Global_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

```python
kf.split(df)
```

```python
from sklearn.linear_model import LinearRegression
lrc = LinearRegression()
from sklearn.preprocessing import PolynomialFeatures
prc = PolynomialFeatures()
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]

    lrc.fit(X_train_kfold,Y_train_kfold)
    predict_lrc = lrc.predict(X_test_kfold)

    lprc = LinearRegression()
    X_train_kfold_prc = prc.fit_transform(X_train_kfold)
    lprc.fit(X_train_kfold_prc,Y_train_kfold)
    X_test_kfold_prc = prc.fit_transform(X_test_kfold)
    predict_prc = lprc.predict(X_test_kfold_prc)

    dtr = dtr.fit(X_train_kfold,Y_train_kfold)
    predict_dtr = dtr.predict(X_test_kfold)
```

```python
print("Набор предсказаний регрессии дерева решений: ",predict_dtr)
print("Набор предсказаний Полиномиальная регрессия: ",predict_prc)
print("Набор предсказаний линейной регрессии: ",predict_lrc)
```
